```python
# loading and reading the CSV files
reads = pd.read_csv("csv-files/reads.csv")
reads.head()
# Value check of reads.csv file,
# Brief is below this code
#for col in reads.columns:
#	print("Values per column: ", reads[col].value_counts())
#* the data shows:
# > dtypes are int64
# > R1-2-3: ICH, IPH, IVH, SDH, EDH, SAH
# > ~: BleedLocation-Left / -Right
# > ~: ChronicBleed, Fracture, CalvarialFracture, OtherFracture
# > ~: MassEffect, MisslineShift
# > ~: No missing values
# > the R1-3 are readers and each reader has 14 features
# "name" is the unique id to check patients
# 491 patients
# B1 is real-world appearance of ICH in a 30 days period	=~ 214 (/) ICH%
# B2 is enriched in ICH (double checked)					=~ 277 (/) ICH%

# iterating rows to find ICH per patient
# 1 is ICH, 0 is Norm
ich_3r = 0
ich_2r = 0
ich_1r = 0
for i, r in reads.iterrows():
	#print(f"Patient {r['name']}/{r['Category']} = R1: {r['R1:ICH']}/R2: {r['R2:ICH']}/R3: {r['R3:ICH']}")
	if r['R1:ICH'] == 1 and r['R2:ICH'] == 1 and r['R3:ICH'] == 1:
		ich_3r += 1
	if r['R1:ICH'] == 1 and r['R2:ICH'] == 1 and r['R3:ICH'] == 0:
		ich_2r += 1
	if r['R1:ICH'] == 1 and r['R2:ICH'] == 0 and r['R3:ICH'] == 1:
		ich_2r += 1
	if r['R1:ICH'] == 0 and r['R2:ICH'] == 1 and r['R3:ICH'] == 1:
		ich_2r += 1
	if r['R1:ICH'] == 1 and r['R2:ICH'] == 0 and r['R3:ICH'] == 0:
		ich_1r += 1
	if r['R1:ICH'] == 0 and r['R2:ICH'] == 0 and r['R3:ICH'] == 1:
		ich_1r += 1
	if r['R1:ICH'] == 0 and r['R2:ICH'] == 1 and r['R3:ICH'] == 0:
		ich_1r += 1
print(f"ICH only 1 Rad: {ich_1r}\nICH only 2 Rad: {ich_2r}\nICH all rads: {ich_3r}")
```

```python
# loading and reading the CSV files
reads = pd.read_csv("csv-files/reads.csv")
reads.head()
# Value check of reads.csv file,
# Brief is below this code
#for col in reads.columns:
#	print("Values per column: ", reads[col].value_counts())
#* the data shows:
# > dtypes are int64
# > R1-2-3: ICH, IPH, IVH, SDH, EDH, SAH
# > ~: BleedLocation-Left / -Right
# > ~: ChronicBleed, Fracture, CalvarialFracture, OtherFracture
# > ~: MassEffect, MisslineShift
# > ~: No missing values
# > the R1-3 are readers and each reader has 14 features
# "name" is the unique id to check patients
# 491 patients
# B1 is real-world appearance of ICH in a 30 days period	=~ 214 (/) ICH%
# B2 is enriched in ICH (double checked)					=~ 277 (/) ICH%
# Agreement of the radiologist differ on the scans:
# Only 1 Rad agrees = 41 scans
# Only 2 Rads agree = 37 scans
# All 3 Rads agree = 168 scans

b1_count = 0
b2_count = 0
for i, r in reads.iterrows():
	# print(f"Patient {r['name']}/{r['Category']} = R1: {r['R1:ICH']}/R2: {r['R2:ICH']}/R3: {r['R3:ICH']}")
	if r['R1:ICH'] == 1 or r['R2:ICH'] == 1 or r['R3:ICH'] == 1:
		if r['Category'] == 'B1':
			b1_count += 1
		if r['Category'] == 'B2':
			b2_count += 1
print(f"B1 ICH: {b1_count}\nB2 ICH: {b2_count}")
```
